**DEEP LEARNING : PYTORCH NEURAL NETWORKS**

```python
# Preparing my data before I convert into tensors
X_train = X_train
X_test = X_test
y_train = y_train.values.reshape(-1,1)
y_test = y_test.values.reshape(-1,1)
```

```python
X_train = torch.from_numpy(X_train).type(torch.Tensor)
X_test = torch.from_numpy(X_test).type(torch.Tensor)
y_train = torch.from_numpy(y_train).type(torch.Tensor)
y_test = torch.from_numpy(y_test).type(torch.Tensor)
```

```python
X_train.shape ,y_train.shape
```

Deep Learning Model
Considering the shape of our target and features I will build a Neural Network with ;
* 29 input features 
* 58 hidden features 
* 1 output features but its multiclass

I will use a Sigmoid Function;
The Sigmoid function performs the role of an activation function in machine learning which is used to add non-linearity in a machine learning model. Basically, the function determines which value to pass as output and what not to pass as output.For my case the output becomes either a 0 or 1.

To get values close to 0 or 1 ,I will be forced to round off my values using **torch.round()** function

```python
class CreditDetector(nn.Module):
    def __init__(self,input_dim=29,hidden_dim=58,output_dim=1,p=0.3):
        super().__init__()
        self.input_dim =input_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.dropout = nn.Dropout(p)
        self.linear1 = nn.Linear(input_dim,hidden_dim)
        self.linear2 = nn.Linear(hidden_dim,hidden_dim)
        self.linear3 = nn.Linear(hidden_dim,hidden_dim)
        self.fc = nn.Linear(hidden_dim,output_dim)
    def forward(self,x:torch.Tensor):
        x = F.relu(x)
        x = torch.sigmoid(x)
        x = self.dropout(self.linear1(x))
        x = self.linear2(x)
        x = self.linear3(x)
        x = self.fc(x)
        x = torch.sigmoid(x)
        x = torch.round(x)
        return x
```

# MODEL TRAINING : 
Before using our model we need to train,the quality of the training data directly impacts the accuracy and reliability of the machine learning algorithm. If the training data is biased or incomplete, the machine learning algorithm will make inaccurate predictions

I will use the following :
*  Adam Optimizer
* Binary CrossEntropyLoss
* TorchMetrics accuracy

```python
# instantiating my model
model = CreditDetector()
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
criterion = nn.BCELoss()
accuracy  = torchmetrics.Accuracy(task='binary')
torch.manual_seed(29112023)
```

MODEL TRAINING

```python
epochs = 100
for epoch in range(epochs):
    y_pred = model.forward(X_train)
    loss = criterion(y_pred,y_train)
    training_accuracy = accuracy(y_pred,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    with torch.no_grad():
        y_pred = model.forward(X_test)
        testing_loss = criterion(y_pred,y_test)
        test_accuracy = accuracy(y_pred,y_test)
        if epoch % 10 == 0 :
            print(f"The epoch is {epoch}|The training loss is {training_loss}|training_accuracy{training_accuracy}| testing_accuracy{test_accuracy}")
        
        
```

```python
# making model predictions
with torch.inference_mode():
    predictions = model.forward(X_train)
    print(f"My first ten predictions are {predictions[:20]}")
```
